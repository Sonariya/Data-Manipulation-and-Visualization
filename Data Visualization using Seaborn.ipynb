```python
### Line plot 
# loading dataset 
data = sns.load_dataset("iris") 
```

```python
data.head()
```

```python
# draw lineplot
sns.lineplot(x='sepal_length',y='sepal_width',data=data)
plt.show()
```

```python
sns.lineplot(x='sepal_length', y='species', data=data)
plt.show()
```

```python
from scipy import stats
np.std(data)
```

Versicolor can be found in range 5.5 units to 7 units

```python
## Scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=data,hue='species')
plt.show()
```

```python
sns.scatterplot(x='petal_length', y='petal_width', data=data,hue='species')
plt.show()
```

with repect  to each class sepal width and sepal lenght both cannot separate all classes.

```python
## Bar plot
##A barplot is basically used to aggregate the categorical data according to some methods and by default its the mean. 
##It can also be understood as a visualization of the group by action. To use this plot we choose a categorical column for 
##the x axis and a numerical column for the y axis and we see that it creates a plot taking a mean per categorical column. 

sns.barplot(x='species', y='sepal_length', data=data)
plt.show()
```

```python
## Count Plot
##A countplot basically counts the categories and returns a count of their occurrences. 
sns.countplot(x='species', data=data)
plt.show()
```

```python
## Box plot
##A boxplot is sometimes known as the box and whisker plot.It shows the distribution of the quantitative data 
## that represents the comparisons between variables. boxplot shows the quartiles of the dataset while the whiskers extend to
## show the rest of the distribution i.e. the dots indicating the presence of outliers.
sns.boxplot(x='sepal_width',data=data)
```

```python
sns.boxplot(x='species', y='sepal_width', data=data)
plt.show()
```

```python
## Violinplot
## It is similar to the boxplot except that it provides a higher, more advanced visualization and uses the kernel 
##density estimation to give a better description about the data distribution.

sns.violinplot(x='species', y='sepal_width', data=data)
plt.show()
```

```python
##Stripplot
## It basically creates a scatter plot based on the category. It is created using the stripplot()
sns.stripplot(x='species', y='petal_width', data=data)
plt.show()
```

```python
## Factorplot
##Factorplot is the most general of all these plots.

sns.factorplot(x='species', y='sepal_width', data=data)
plt.show()
```

```python
## Histogram
## A histogram is basically used to represent data provided in a form of some groups.
## It is accurate method for the graphical representation of numerical data distribution
sns.histplot(x='sepal_width', data=data)
plt.show()
```

```python
sns.histplot(x='sepal_width',hue='species',data=data)
plt.show()
```

```python
## Distplot
## Distplot is used basically for univariant set of observations and visualizes it through a histogram 
## i.e. only one observation and hence we choose one particular column of the dataset
sns.distplot(data['sepal_width'])
plt.show()
```

```python
##Pairplot
##Pairplot represents pairwise relation across the entire dataframe and supports an additional 
## argument called hue for categorical separation. What it does basically is create a jointplot between every possible numerical 
##column and takes a while if the dataframe is really huge

sns.pairplot(data=data, hue='species')
plt.show()
```

```python
##Heat map
## Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix.
## In this, to represent more common values or higher activities brighter colors basically reddish colors are used and 
## to represent less common or activity values, darker colors are preferred

# correlation between the different parameters 
tc = data.corr()
  
sns.heatmap(tc)
plt.show()
```
