```python
# multiline plot
a=data['facecream']
b=data['facewash']
c=data['toothpaste']
d=data['bathingsoap']
e=data['moisturizer']
f=data['shampoo']
month=data['month_number']


plt.plot(month, a,label = 'Face cream Sales', marker='o')
plt.plot(month, b,label = 'Face Wash Sales',  marker='o')
plt.plot(month, c,label = 'ToothPaste Sales', marker='o')
plt.plot(month, d,label = 'Bathingshop Sales', marker='o')
plt.plot(month, f,label = 'Shampoo Sales', marker='o')
plt.plot(month, e,label = 'Moisturizer Sales', marker='o')

plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.xticks(month)
plt.yticks([1000, 2000, 4000, 6000, 8000, 10000, 12000, 15000, 18000])
plt.title('Sales data')
plt.show()
# in this graph we can see that bathing shop sales and toothpaste sales increasing in every month proportionally
#facecream sales rapidly decreses after 9th month and after 10th month it is increasing proportionally with month again
#Shampoo sales has highest sales during 3rd  month and 8th month
# moisturizer sales is decreasing proportionally after 11th month
```

```python
## Q5)Read toothpaste sales data of each month and show it using a scatter plot
sns.scatterplot(x='month_number',y='toothpaste',data=data)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Toothpaste Sales Data')
plt.show()
# in this graph we can see toothpaste has highest sales on 10th month
```

```python
#Q6)Read face cream and facewash product sales data and show it using the bar chart
month=data['month_number']
faceCremSales=data['facecream']
faceWashSales=data['facewash']

plt.bar([a-0.25 for a in month],faceCremSales,label = 'FaceCream sales',width= 0.25,align='edge') 
## Passing the parameters to the bar function, this is the main function which creates the bar plot
# Using 'a' now to align the bars side by side
plt.bar([a+0.25 for a in month],faceWashSales,label = 'FaceWash sales',width= -0.25,align='edge')
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Sales data')

plt.xticks(month)
plt.grid(True, linewidth= 1, linestyle="--")
plt.title('Facewash and facecream sales data')
plt.show()
```

```python
#Q7)Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk
month = data['month_number']
bathingsoapSales= data['bathingsoap']
plt.bar(month,bathingsoapSales)
plt.xlabel('Month Number')
plt.ylabel('Bathingsoap Sales')
plt.xticks(month) ## is used to get and set the current tick locations and labels of the x-axis
plt.grid(True, linewidth= 1, linestyle="-.",mec='c')
plt.title('Bathingsoap Sales Data')
plt.savefig('C:/Users/Ariya/sales_data_of_bathingsoap.png', dpi=150)
plt.show()
```

```python
##8>Read the total profit of each month and show it using the histogram to see the most common profit ranges
profit = data['total_profit']
profit_range = [150000, 200000, 250000, 300000, 350000,450000]
plt.hist(profit, profit_range)
plt.xlabel('profit range')
plt.ylabel('Actual Profit')
plt.xticks(profit_range)
plt.title('Profit data')
plt.show()
```

```python
##9>Calculate total sale data for last year for each product and show it using a Pie chart

month= data['month_number']

labels = ['FaceCream', 'FaseWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
salesData   = [data['facecream'].sum(), data['facewash'].sum(), data['toothpaste'].sum(), 
         data['bathingsoap'].sum(), data['shampoo'].sum(), data['moisturizer'].sum()]
plt.axis("equal")
plt.pie(salesData, labels=labels, autopct='%.1f%%')
plt.title('Sales data')
plt.show()
```

```python
##10>Read Bathing soap facewash of all months and display it using the Subplot
month= data['month_number']
bathingsoapSales= data['bathingsoap']
faceWashSales=data['facewash']

fig,a = plt.subplots(2, sharex=True) # sharex means sharing x axis for both plot
a[0].plot(month, bathingsoapSales, label = 'Bathingsoap Sales Data', color='k', marker='o', linewidth=3)
a[0].set_title('Sales data of  a Bathingsoap')
a[1].plot(month, faceWashSales, label = 'Face Wash Sales Data', color='r', marker='o', linewidth=3)
a[1].set_title('Sales data of  a facewash')

plt.xticks(month)
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.show()
# in this graph we can see that Bathingsoap sales is incresing ather 9th  month proportionally
# whereas sales of facewash is decresing after 11th month drastically
```
