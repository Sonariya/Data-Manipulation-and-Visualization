## line Plot
- used to plot numerical continuous data

```python
x=[1,8,9,5,3,4,7,5,6,8,12]
y=[1,4,3,6,7,3,6,7,8,9,13]
```

```python
plt.plot(x,color='m')
plt.show()
```

```python
plt.plot(y,color='m',linestyle='-.',marker='o')
plt.show()
```

```python
## plotting two continous variables in line plot
plt.plot(x,y,color='c',marker='o',linestyle='--') ## relationship
plt.show()
```

```python
## two variables in same plot
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
x=np.array([[1,2,3],[4,5,6]])
plt.plot(x)
plt.show()
```

## Scatter Plot
It will always give relationship between two continous variables

```python
x=[1,8,9,5,3,4,7,5,6,8,12]
y=[1,4,3,6,7,3,6,7,8,9,13]
```

```python
plt.scatter(x,y)
plt.show()
```

## Bar Plot
when we want to draw discrete data

```python
x=[2,8,10]
y=[11,16,9]
x1=[3,9,10]
y2=[11,16,12]
```

```python
plt.bar(x,y,color='c')
plt.bar(x1,y2)
plt.show()
```

## Histogram
First ir will create bins or interval in data and then it will count the samples which are presents in
the corresponding bin.
X axis--> represents bins
Y axis-->represents frequency /count of data in given samples

```python
x=np.random.normal(170,10,250)
x
```

```python
plt.hist(x)
plt.show()
```

## Pieplot
Use it for categorical and discrete data

```python
pielist=[1,2,3,4,5]
plt.pie(pielist,labels=['Mango','Banana','Orange','Kiwi','Pomogranate'])
plt.show()
```

```python
labels=['Python','C++','Java','ASP.Net']
sizes=[215,130,245,210]
color=['gold','yellowgreen','cyan','lightcoral']
explode=[0.2,0,0,0.1]
plt.pie(sizes,labels=labels,colors=color,explode=explode,autopct="%.2f")
plt.show()
```

```python
## object oriented plotting
import math
x=np.arange(0,math.pi*2,0.05)
x
```

```python
y=np.sin(x)
# creating Blank or empty canvas
fig=plt.figure()
## Adding axes
ax=fig.add_axes([0,0,1,1])
# Labeling the axes
ax.set_title('Sine Wave')
ax.set_xlabel('Angle')
ax.set_ylabel('Sine')
ax.plot(x,y)
plt.show()
```

```python
## Common layout plots
fig,a=plt.subplots(2,2)
x=np.arange(1,6)
a[0][0].plot(x,x*x)
a[0][0].set_title('First Plot')
a[0][1].hist(x)
a[0][1].set_title('Histogram')
a[1][0].plot(x,np.sqrt(x))
a[1][0].set_title('Square root')
a[1][1].plot(x,np.log(x))
a[1][1].set_title('Log')
plt.show()
```
